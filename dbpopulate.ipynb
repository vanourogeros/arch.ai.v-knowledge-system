```python
import requests

def fetch_articles(api_key, category, num_articles=10):
    try:
        # Make a request to Diffbot's Article API to fetch articles based on category
        response = requests.get(f"https://api.diffbot.com/v3/search?token={api_key}&query=type:article&tags={category}&size={num_articles}")
        data = response.json()
        print(data)

        articles = data.get('objects', [])
        for article in articles:
            article_url = article.get('pageUrl')
            full_text = fetch_full_content(api_key, article_url)
            print("Full Text of Article:", full_text)
    except Exception as e:
        print("An error occurred while fetching articles:", str(e))

def fetch_full_content(api_key, article_url):
    try:
        # Make a request to Diffbot Article API
        response = requests.get(f"https://api.diffbot.com/v3/article?token={api_key}&url={article_url}")
        data = response.json()
        
        # Extract full text from the response
        full_text = data.get('objects', [{}])[0].get('text', 'N/A')
        return full_text.strip()
    except Exception as e:
        print("Failed to fetch full content:", str(e))
        return 'N/A'

# Replace 'YOUR_API_KEY' with your actual Diffbot API key
api_key = '81a2ca949312724123e21bbe7ecb9af2'

# Specify the category of articles you want to fetch
category = 'technology'

# Number of articles to fetch
num_articles = 10

fetch_articles(api_key, category, num_articles)
```
